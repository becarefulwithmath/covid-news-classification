```python
#!/us bbbr/bin/env python

#for search
from googlesearch import search
import csv
import random
import datetime
query_suffix = " (fallacy OR bias OR heuristic) (coronavirus OR covid OR pandemic)"
number_or_results = 10
#add article text1
from newspaper import Article

#add article text2
from bs4 import BeautifulSoup
from bs4.element import Comment
import urllib.request
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True
def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)

#file creation
now = datetime.datetime.now()
name = now.strftime("%m%d%Y%H%M%S")
file_name = "output_google_no_text" + name + ".csv"
print(file_name) 
with open(file_name, "a", newline='') as w:
            writer = csv.writer(w)
            row = ["query_id", "query", "result position", "url"]
            print(row)
            writer.writerow(row)

#array with search queries
array_of_q = []
with open("Google keywords (biases).csv", "r") as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        array_of_q.append(row[1]+query_suffix) 
print(array_of_q)

#search and results write
num_id = 1
for i in array_of_q:
    num = 0
    for url in search(i, tld='com', lang='en', stop=number_or_results, pause=random.randint(1,3)):
        num += 1
        with open(file_name, "a", newline='', encoding='utf-8') as w:
            writer = csv.writer(w)
                #write results
            row = [str(num_id), str(i), str(num), str(url)]
            #print(row)
            writer.writerow(row)
    num_id += 1
    print(str(len(array_of_q)+1-(num_id))+' remained')
```
